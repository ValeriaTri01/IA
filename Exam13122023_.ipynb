```python
df.head()
del df['sepal_length']
del df['sepal_width']
```

```python
df.head(5)
```

```python
df ['species'].unique()
```

```python
df ['target']=df ['species'].map({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3 })
```

```python
df.head(5)
```

```python
y=df['target']
```

```python
x=df.drop(['target','species'],axis=1)
df.head(5)
```

```python
type(x)
x.shape
import numpy as np
x=x.to_numpy()
print(type(x))
```

split the data set into

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,xtest=train_test_split(x,y,test_size=0.3, random_state=240 )
```

create the model of classification

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression ()
```

training step

```python
model.fit(xtrain,ytrain)
```

```python
predictions=model.predict(xtest)
```

```python
print(predictions[:10])
```

```python
print(ytest[:10])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true=ytest,y_pred=predictions)
```

```python
import pickle
pickle.dump(model,open('flores.pkl','wb'))
```
