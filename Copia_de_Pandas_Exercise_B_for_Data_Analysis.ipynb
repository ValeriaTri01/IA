# 1. Import pandas as pd and load the dataset

```python
# Use this line to load the dataset.
import pandas as pd
url='https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/Station_Entries_Daily_Totals.csv'
# your code here
df = pd.read_csv(url)
print (df.head())
pass
```

Basic Exploration

Let's first take a quick look at what this data looks like.

```python
# for basic exploration, we suggest to use shape, head(), tail(), dtypes, type() modules
# your code here
# Display basic information
print("Dataset Shape:", df.shape)
print("\nDataset Head:")
print(df.head(5)) # primera fila
print("\nDataset Tail:")
print(df.tail()) #las utimas 5 filas
print("\nData Types:")
print(df.dtypes)
pass
```

What are some basic statistics about the data?

```python
# use the describe module utilizar el módulo de descripción
# your code here
print("Descriptive Statistics:")
print(df.describe())
pass
```

Selection Based on Conditionals

Let's see which stations have the most and fewest rides.


```python
# use the min() and max() modules
# your code here

# Find the station with the most rides # Encuentra la estación con más viajes

print("\Maximo:")
print(df['rides'].idxmax())

# Find the station with the fewest rides # Encuentra la estación con la menor cantidad de viajes.

print("\Minimo:")
print(df['rides'].idxmin())

print
pass
```

Create a Time Index

So we have all rides from January 1, 2001 to December 31, 2019. Let's do some statistics that group ride statistics by date. First we need to tell Pandas that the date column is in fact a date. So, we convert the column to a proper 'DateTime' type, and then set the index to this column.

This step takes a little bit of time!

The following picture show the desired output

image.png

```python
# hint, use to_datetime() and set_index()  modules
# your code here


# Convert the 'date' column to datetime Convertir la columna 'fecha' a fecha y hora
df['date'] = pd.to_datetime(df['date'])

# Set the 'date' column as the index Establecer la columna 'fecha' como índice
df.set_index('date', inplace=True)

# Display the resulting DataFrame with the time index   Mostrar el DataFrame resultante con el índice de tiempo
print(df.head())
pass
```

Sort the Dataframe by Date

```python
# hint, sort_values(by='date')
# your code here


# Sort the DataFrame by date  Ordenar el DataFrame por fecha
df.sort_values(by='date', inplace=True)


print(df.head())
pass
```

Sanity Checking

Looks good! Now let's have a quick look at data for specific stations:
'Morgan-Lake' and 'Addison-Brown'


```python
# hint, use filter like, df['stationname']=='Morgan-Lake']
# your code here
# use the following variable name for your specific stations
#Morgan_Lake=pd.DataFrame()


# Filtrar datos para 'Morgan-Lake' y 'Addison-Brown'
Morgan_Lake = df[df['stationname'] == 'Morgan-Lake']


# Mostrar datos para 'Morgan-Lake'
print("\nDatos para 'Morgan-Lake':")
print(Morgan_Lake.head())

pass
```

```python
# hint, use filter like this df['stationname']=='Addison-Brown']
# your code here
# use the following variable name for your specific stations
#Addison_Brown=pd.DataFrame()




# Filtrar datos para 'Morgan-Lake' y 'Addison-Brown'

Addison_Brown = df[df['stationname'] == 'Addison-Brown']



# Mostrar datos para 'Addison-Brown'
print("\nDatos para 'Addison-Brown':")
print(Addison_Brown.head())
pass
```

Visualizing Timeseries Data

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(11, 4)})
# Morgan_Lake['rides'].plot(linewidth=0.5)

# Establecer el estilo y el tamaño de la figura de Seaborn
sns.set(rc={'figure.figsize':(11, 4)})

# Graficar los datos de 'rides' para 'Morgan-Lake' usando Pandas
Morgan_Lake['rides'].plot(linewidth=0.5)

# Personalizar el gráfico
plt.title('Viajes a lo largo del tiempo para la estación Morgan-Lake')
plt.xlabel('Fecha')
plt.ylabel('Número de Viajes')
plt.show()


pass
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
# Graficar los datos de 'rides' para 'Addison-Brown' usando Pandas
Addison_Brown['rides'].plot(linewidth=0.5)

# Personalizar el gráfico
plt.title('Viajes para la estación de  Addison-Brown')
plt.xlabel('Fecha')
plt.ylabel('Número de Viajes')
plt.show()
pass
```
